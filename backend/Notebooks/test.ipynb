```python
def clean_text(page_text: str):
    remove_list = [
        " (en)", "[réf. nécessaire]"
    ]
``
    replace_list = {
        "(": ", ",
        ")": ", "
    }
    cleaned_text = page_text

    for item_to_remove in remove_list:
        cleaned_text = cleaned_text.replace(item_to_remove, "")
    
    for to_replace, replacer in replace_list.items():
        cleaned_text = cleaned_text.replace(to_replace, replacer)

    return cleaned_text

test_text = "jazz (en), le soul jazz, le jazz de chambre ('crazy jazz'), le latin jazz, le jazz-funk, le loft jazz (en), le jazz punk, l'acid jazz, le jazz rap, le M-Base (en) et le nu jazz."
clean_text(test_text)
```
